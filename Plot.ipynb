```python
## Creating a plot from numpy array!
```

```python
import numpy as np
```

```python
x = np.array([12,15,30,50])
y = np.array([20,40,60,80])
```

```python
x
```

```python
plt.plot(x,y, 'r--')
plt.show()
```

```python
## How to add another line to the figure!
```

```python
plt.plot(x,y, 'r--', b, a, "g^")
plt.show()
```

```python
plt.plot(x,y, 'r--', b, a, "g^", x, y**2, 'bs')
plt.show()
```

 ### Scater Plot!

```python
height = [10, 40, 20, 67, 131,11]
weight = [75.7, 80.2, 62.5,73, 68.2, 100]
```

```python
plt.scatter(height, weight, c= 'violet')
plt.show()
```

```python
plt.scatter(height, weight, color= 'violet', s = height)
plt.show()
```

### Bar Chat
we use bar Chat to plot Categorical varriables

```python
sex = ["boy", "girl"]
Age = [12, 24]
```

```python
plt.bar(sex, Age, color= 'black')
plt.show()
```

```python
plt.barh(sex, Age, color= 'black')
plt.show()
```

```python
plt.barh(sex, Age, color= 'black')
plt.ylabel("sex")
plt.xlabel("Age")
plt.show()
```

```python
### Plotting a 2 by 2 dimensional figure(it will contain 4 charts)
```

```python
plt.figure(figsize= (15,5))
plt.subplot(2,2,1)
plt.plot(x,y, 'r--', b, a, "g^", x, y**2, 'bs')
plt.subplot(2,2,2)
plt.scatter(height, weight, color= 'violet', s = height)
plt.subplot(2,2,3)
plt.bar(sex, Age, color= 'black')
plt.subplot(2,2,4)
plt.barh(sex, Age, color= 'black')
plt.ylabel("sex")
plt.xlabel("Age")
plt.suptitle("A two by two dimensional plot")
plt.show()
```
