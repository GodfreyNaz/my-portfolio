```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("gettyimages-2068694733.jpg")
```

```python
img
```

```python
# convert images to greyscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray
```

```python
plt.imshow(img_gray, cmap= "gray")
plt.show()
```

```python
# Convert greyscale image to array
img_arr = np.array(img_gray)
```

```python
img_arr
```

```python
img_arr.shape
```

```python
## How to resize an image
cv2.resize(img, (65, 64))
```
