```python
def load_dataset_from_images(floder, label, img_size = (65,65)):
    images =[]
    labels= []
    for filename in os.listdir(Pineapple_folder):
        img_path = os.path.join(Pineapple_folder, filename)
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        img_resize = cv2.resize(img, img_size)
        img_flattened = img_resize.flatten()
        images.append(img_flattened)
        labels.append(label)
    return images, labels
```

```python
Pineapple_images, Pineapple_labels = load_dataset_from_images(Pineapple_folder, 0)
Pineapple_Mini_images, Pineapple_Mini_labels = load_dataset_from_images(Pineapple_Mini_folder, 1)
Raspberry_images, Raspberry_labels = load_dataset_from_images(Raspberry_folder, 2)
Redcurrant_images, Redcurrant_labels = load_dataset_from_images(Redcurrant_folder, 3)
Strawberry_Wedge_images, Strawberry_Wedge_labels= load_dataset_from_images(Strawberry_Wedge_folder, 4)
```

```python
X = np.array(Pineapple_images + Pineapple_Mini_images + Raspberry_images + Redcurrant_images + Strawberry_Wedge_images)
Y = np.array(Pineapple_labels + Pineapple_Mini_labels + Raspberry_labels + Redcurrant_labels + Strawberry_Wedge_labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, random_state = 10, test_size = 0.2)
```

```python
from sklearn.svm import SVC
```

```python
SVM = SVC(random_state= 20)
```

```python
SVM.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
predicted_values = SVM.predict(x_test)
```

```python
print(classification_report(y_test, predicted_values))
```
