```python
def face_detection(image):
    #read the image 
    img = cv2.imread(image)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    x,y,w,h = faces[0]
    face_img = cv2.rectangle(img, (x,y), (x + w, y + h), (0,255,255), thickness = 8)
    return plt.imshow(face_img)
```

```python
## take in video capture
cap = cv2.VideoCapture(0)
```

```python
while True:
    result, frame = cap.read()
    if result is False:
        break
    img_grey= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier("haarcasc  ade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(img_grey, 1.3,5)
    for x,y,w,h in faces:
        cv2.rectangle(frame, (x,y), (x + w, y + h), (0,255,255), 4)
        cv2.imshow("Face Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
