```python
Wall.fillna({"Branch": "A", "City": "Yangon", "Gender": "female", "Payment": "Ewallet"}, inplace = True)
Wall.fillna({"Unit price": 55.61, "Quantity": 5.5, "Rating": 6.9,}, inplace = True) 
```

```python
(Wall.isnull().sum())/1000*100
```

```python
Wall
```

```python
# Create a new Coloum for Gross Revenue
```

```python
Wall["Gross Revenue"] = Wall["Unit price"] * Wall["Quantity"]
```

```python
### Create a column for Gross profit
```

```python
Wall["Gross Profit"] = Wall["Gross Revenue"] - Wall["cogs"]
```

### What month had the Hightes Cost of Goods sold

```python
### Month with the largest Sales
```

```python
MT = Wall.groupby("Month").agg({"cogs": "sum"}).sort_values(by = ["cogs"], ascending = False).reset_index()
```

```python
MT
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(MT['Month'], MT['cogs'], color='Black')
plt.xlabel('Months')  # Label for x-axis
plt.ylabel('Max cogs')   # Label for y-axis
plt.title('Month With The Largest Cost Of Good Sold')
plt.show()
```

```python
### City with the Largest Revenue
```

```python
CTE= Wall.groupby("City").agg({"Gross Revenue": "max"}).reset_index()
CTE
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(CTE['City'], CTE['Gross Revenue'], color='Black')
plt.xlabel('Cities')  # Label for x-axis
plt.ylabel('Max Gross Revenue')   # Label for y-axis
plt.title('City with the Larget Revenue')
plt.show()
```

```python
### Common Product line by Gender
```

```python
MPL= Wall.groupby("Gender").agg({"Product line": "count"}).reset_index(["Gender"])
MPL
```

```python
# Pivot the data for a stacked bar chart
MPL = PL.pivot(index='Product line', columns='Gender', values='Count').fillna(0)

# Plotting a stacked bar chart
MPL.plot(kind='bar', stacked=True, figsize=(8, 5))

plt.xlabel('Product Line')
plt.ylabel('Count')
plt.title('Count of Product Lines by Gender (Stacked)')
plt.legend(title='Gender')
plt.show()
```
