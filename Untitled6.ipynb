```python
## Principal component Analysis (PCA).. this is used for reducing the dimension of your input to just 2 coloumns
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_table("fruit_data_with_colors.txt")
```

```python
df.head()
```

```python
x = df.drop(["fruit_name","fruit_name", "fruit_subtype"], axis = 1)
```

```python
x.head()
```

```python
plt.scatter(df.mass, df.width)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(x)
```

```python
X
```

```python
pca = PCA(n_components= 2)
```

```python
pca.fit_transform(X)
```

```python
x_pca = pca.fit_transform(X)
```

```python
x_pca[:,0]
```

```python
first_component = x_pca[:,0]
second_component = x_pca[:,1]
```

```python
plt.scatter(first_component, second_component, s= 65, c= df.fruit_label, alpha = 0.8)
plt.xlabel("First Component")
plt.ylabel("Second Component")
plt.show()
```

```python
df.fruit_label.unique()
```

##### K Means Clustering Alogrithm


```python
## This is an unsupervised  which grouped the unlabled dataset into different clustered
```

```python
from sklearn.cluster import KMeans
```

```python
## initialize kmeans
KM = KMeans(n_clusters= 4, random_state = 30)
```

```python
KM.fit(x_pca)
```

```python
KM.labels_
```

```python
plt.scatter(first_component, second_component, s= 65, c= KM.labels_, alpha = 0.8)
plt.xlabel("First Component")
plt.ylabel("Second Component")
plt.show()
```

```python
for i in range (2, 10):
    KM = KMeans(n_clusters= i, random_state = 30)
    KM.fit(x_pca)
    plt.scatter(first_component, second_component, s= 65, c= KM.labels_, alpha = 0.8)
    plt.xlabel("First Component")
    plt.ylabel("Second Component")
    plt.show()
```
