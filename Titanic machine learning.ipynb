```python
x = df.drop("Survived", axis = 1)
y = df["Survived"]
```

```python
x.columns = x.columns.astype(str)
x_columns = x.columns
```

```python
x_columns
```

```python
from sklearn.preprocessing import MinMaxScaler
sk = MinMaxScaler()
x = sk.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 50)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
knn = KNN(n_neighbors = 9)
```

```python
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import f1_score
```

```python
# testing our traning model and checking for accuracy
predicted_output = knn.predict(x_test)
accuracy = f1_score(predicted_output, y_test)
accuracy
```

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
MSE(predicted_output, y_test)
```

```python
accuracy_list = []
k_values = []
for i in range (1,100):
    lnn = KNN(n_neighbors = i)
    lnn.fit(x_train, y_train)
    predicted_output = lnn.predict(x_test)
    accuracy = f1_score(predicted_output, y_test)
    k_values.append(i)
    accuracy_list.append(accuracy)
```

```python
accuracy_list
```

```python
plt.plot(k_values, accuracy_list)
plt.xlabel("k_values")
plt.ylabel("accuracy_list")
plt.show()
```
