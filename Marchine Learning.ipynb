### Label Encoder

```python
df.fruit_name.unique()
```

```python
df = pd.read_table("fruit_data_with_colors.txt")
```

```python
df.head()
```

```python
## Label Encoding
Mydictionary ={"apple": 0, "mandarin": 1, "orange": 2, "lemon": 3}
df["fruit_name"].map(Mydictionary)
```

```python
df["fruit_name"] = df["fruit_name"].map(Mydictionary)
```

```python
df.head()
```

```python
X= df[["mass", "width", "height", "color_score"]]
Y= df["fruit_name"]
```

```python
X
```

```python
Y
```

### K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
###70 % for trainng set 30% for test_set!
###70% for by 30% for x
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 20)
```

```python
X_test.shape
```

```python
### Initialising the KNN classifier

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn
```

```python
## Fit the model on Traning Set

knn.fit(X_train, Y_train)
```

```python
### Checking for accuracy score

knn.score(X_test, Y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors= 7)
knn.fit(X_train, Y_train)
knn.score(X_test, Y_test)
```

```python
n_neighbors = []
traning_accuracy = []
test_accuracy = []
range(10)
```

```python
for i in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(X_train, Y_train)
    b = knn.score(X_train, Y_train)
    a = knn.score(X_test, Y_test)
    traning_accuracy.append(b)
    test_accuracy.append(a)
    n_neighbors.append(i)
```
